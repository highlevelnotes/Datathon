```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 원핫 인코딩 하고난 다음 시각화

df_one_hot = pd.get_dummies(df)

fig, axs = plt.subplots(5, 5, figsize=(30, 30))

for i, column in enumerate(df_one_hot.drop('fraud_bool', axis=1).select_dtypes('number').columns):
    row = i // 5
    col = i % 5
    sns.kdeplot(data=df_one_hot[df_one_hot['fraud_bool'] == 1][column], label='Fraud', fill=True, ax=axs[row, col])
    sns.kdeplot(data=df_one_hot[df_one_hot['fraud_bool'] == 0][column], label='Not-Fraud', fill=True, ax=axs[row, col])
```

```python
# 각 컬럼들의 상관관계를 보자
sns.heatmap(df_one_hot.corr())
```

```python
# 위 상관관계를 통해 proposed_credit_limit (신청자의 신용 한계. [200, 2000]의 범위) / credit_risk_score 높은 것만 밀도 다시 보기

# 밀도 그래프 그리기
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Proposed Credit Limit 밀도 그래프
sns.kdeplot(data=df_one_hot[df_one_hot['fraud_bool'] == 1]['proposed_credit_limit'],
             label='Fraud', fill=True, color='orange', ax=axs[0])
sns.kdeplot(data=df_one_hot[df_one_hot['fraud_bool'] == 0]['proposed_credit_limit'],
             label='Not-Fraud', fill=True, color='blue', ax=axs[0])
axs[0].set_title('Proposed Credit Limit Density Plot')
axs[0].set_xlabel('Proposed Credit Limit')
axs[0].set_ylabel('Density')
axs[0].legend()

# Credit Risk Score 밀도 그래프
sns.kdeplot(data=df_one_hot[df_one_hot['fraud_bool'] == 1]['credit_risk_score'],
             label='Fraud', fill=True, color='orange', ax=axs[1])
sns.kdeplot(data=df_one_hot[df_one_hot['fraud_bool'] == 0]['credit_risk_score'],
             label='Not-Fraud', fill=True, color='blue', ax=axs[1])
axs[1].set_title('Credit Risk Score Density Plot')
axs[1].set_xlabel('Credit Risk Score')
axs[1].set_ylabel('Density')
axs[1].legend()

plt.tight_layout()
plt.show()


#### 사기데이터 !!!  크레딧 리스크 점수는 높고, 신청자의 신용한계는 낮다.
```

```python
##
# _importance() 함수 사용이 가능한, 모델 중 우리가 선정한 모델은

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import precision_score, recall_score, f1_score
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

models = [
    ("Random Forest", RandomForestClassifier(random_state=0)),
    ("Gradient Boosting", GradientBoostingClassifier(random_state=0)),
    ('LightGBM', LGBMClassifier(boost_from_average=False)),
    ('XGBoost', XGBClassifier())
]

```

```python
base_df = df_one_hot

X = base_df.drop('fraud_bool', axis=1)
y = base_df['fraud_bool']

X_train = X[X['month'] < 6]
y_train = y[X['month'] < 6]
X_test = X[X['month'] >= 6]
y_test = y[X['month'] >= 6]

X_train.drop('month', axis=1, inplace=True)
X_test.drop('month', axis=1, inplace=True)
```

### LightGBM 의 피쳐 중요도

```python
model = LGBMClassifier( colsample_bytree=0.8, max_depth=-1, n_estimators=300, num_leaves=50, subsample=0.8, boost_from_average=False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
import lightgbm as lgb

lgb.plot_importance(model, importance_type='split', max_num_features=20)
```

### XGBoost의 피쳐 중요도

```python
import xgboost as xgb


#model = XGBClassifier(tree_method='gpu_hist', gpu_id=0, scale_pos_weight=89.67005)
model = xgb.XGBClassifier(tree_method='gpu_hist', gpu_id=0, scale_pos_weight=89.67005)

model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=True)
y_pred = model.predict(X_test)

```

```python
from xgboost import plot_importance


# fig, axs = plt.subplots(1, 1, figsize=(20, 10))

plot_importance(model, max_num_features=20)

```

### Random Forest 의 피쳐 중요도

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')


# 랜덤 포레스트 학습 및 별도의 테스트 셋으로 예측 성능 평가
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train , y_train)
pred = rf_clf.predict(X_test)
```

```python
ftr_importances_values = rf_clf.feature_importances_
ftr_importances = pd.Series(ftr_importances_values,index=X_train.columns  )
ftr_top20 = ftr_importances.sort_values(ascending=False)[:20]

plt.figure(figsize=(8,6))
plt.title('Feature importances Top 20')
sns.barplot(x=ftr_top20 , y = ftr_top20.index)
plt.show()
```
